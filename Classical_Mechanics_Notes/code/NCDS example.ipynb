```python
x,y = var('x y')
k = 5
s = streamline_plot((y,-k*x),(x,-50,50), (y,-50,50))
p = point((0,0),size = 150, color = 'red')
g = s+p
g.axes_labels(['x','v(x)'])
g.show()
g.save('harmo.png')
```

```python
p1 = plot(0.5*k*x^2, (x, -50, 50),
          axes_labels=['x', 'U(x)'], fontsize=10)
p1.show()
p1.save('Oscill.png')
```
